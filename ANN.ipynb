```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,explained_variance_scor
```

```python
diamonds_train_dummies = pd.read_csv("output/diamonds_train_dummies.csv")
diamonds_train_dummies.head()
```

```python
columnas = [a for a in diamonds_train_dummies.columns if a not in ["price"]]
X = diamonds_train_dummies[columnas]
y = diamonds_train_dummies["price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# define base model

def baseline_model():
    # create model
    model = Sequential()
    # add 1st layer
    model.add(Dense(units=18, input_dim=11, activation='relu')) # kernel_initializer='normal',
    # add hidden layer
    model.add(Dense(units=12, kernel_initializer='normal', activation='relu'))
    # add output layer
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
estimator = KerasRegressor(build_fn=baseline_model, epochs=10, batch_size=5)

kf = KFold(n_splits=5)
results = cross_val_score(estimator, X_train, y_train, cv=kf)
```

```python
print('Results: %.2f (%.2f) MSE' % (results.mean(), results.std()))
```
